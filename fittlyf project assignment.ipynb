### 2.	In column ‘B’ replace Jan with 1, feb with 2, march with 3 and so on…

```python
df['Month'].unique()
```

```python
def data_cleaning(df):
    df=df.fillna(0)
    
    monthdict={'Jan':1,'Feb':2,'Mar':3,'Apr':4, 'May':5, 'Jun':6, 'Jul':7, 'Aug':8, 'Sep':9,'Oct':10,'Nov':11,'Dec':12}

    def month_to_num(month):
        return monthdict[month]
    df['Month'] = df['Month'].apply(month_to_num)
    df['Where_Are_They_comming_from?']=df['Where_Are_They_comming_from?'].replace('Came_From_Google','Google')
    df['Where_Are_They_comming_from?']=df['Where_Are_They_comming_from?'].replace('Landed_on_the_page_Directly','Direct_traffic')
    return df
```

```python
new_data=data_cleaning(df)
new_data
```

## Part 4
Simple Machine learning questions

Write a function called pred_future() which, when called, would perform the following activity:

Predict “How_many_Landed_on_the_our_Page_and_clicked_on_a_button_and_started_f illing_the_Form_and_Completed_and_submited_the_form?” for the complete year of 2022
Generate the overall MAPE of your prediction for the year 2021.

```python
def  pred_future(fb): 
        from sklearn.model_selection import train_test_split

        y = df[["How_many_Landed_on_the_our_Page_and_clicked_on_a_button_and_started_filling_the_Form_and_Completed_and_submited_the_form?"]]
        X = df[['How_many_Landed_on_the_our_Page_and_clicked_on_a_button_and_started_filling_the_Form?']]
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 1)

        from sklearn.linear_model import LinearRegression
        regressor = LinearRegression()
        regressor.fit(X_train, y_train)

        y_pred = regressor.predict(X_test)
        return y_pred
pred_future(df)
        
df['predicted_values']=pd.DataFrame(pred_future(df))
```
